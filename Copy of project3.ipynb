```python
days=[1,2,3,4,5,6,7,8]
speed=[60,62,61,58,56,57,46,63]
plt.plot(days,speed,"--",marker="+",color="g",label="CSD")
plt.scatter(days,speed,c="b",s=45)
plt.title("car speed measurement")
plt.legend()
plt.xlabel("days")
plt.ylabel("car speed")
plt.show()
```

```python
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]
plt.plot(days,max_speed,"r",label="Maximum Speed")
plt.plot(days,min_speed,"g",label="Minimum Speed")
plt.plot(days,avg_speed,"b",label="Average Speed")
plt.title("car speed measurement")
plt.xlabel("days")
plt.ylabel("car speed")
plt.legend()
plt.grid()
plt.show()
```
